```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("victims.csv")
df.head(20)
```

```python
df = df[df.level_1.str.contains("Total")].reset_index(drop=True)
df.head(50)
```

```python
new_value = df.value.groupby(df.value.index // 4).sum().reset_index(drop=True)
print(new_value.to_string())
```

```python
df = df.drop(columns=["level_2"]).drop_duplicates(subset=["year", "level_1"]).reset_index(drop=True)
df.value = new_value
df.head(20)
```

```python
offences = ["Murder", "Rape", "Outrage Of Modesty", "Robbery", "Snatch Theft", "Rioting", "Serious Hurt", "Cheating & Related"]

for offence in offences:
    df.loc[df.level_1.str.contains(offence), "level_1"] = offence

df.rename(columns={"level_1":"target"}, inplace=True)

df.head(20)
```

```python
years = df.year.unique()

for year in years:
    year_df = df[df.year == year].drop(columns=["year"])
    year_df["source"] = pd.Series(["Non-Sexual",
                                "Sexual",
                                "Sexual",
                                "Non-Violent",
                                "Non-Violent",
                                "Non-Sexual",
                                "Non-Sexual",
                                "Non-Violent"], index=year_df.index)
    sexual_value = year_df[year_df.source.str.match("Sexual")].value.sum()
    nonsexual_value = year_df[year_df.source.str.match("Non-Sexual")].value.sum()
    new_rows = pd.DataFrame.from_dict({"source": ["Violent", "Violent"],
                                       "target": ["Sexual", "Non-Sexual"],
                                       "value": [sexual_value, nonsexual_value]})
    year_df = year_df.append(new_rows, ignore_index=True)
    cols = year_df.columns.tolist()
    year_df = year_df[cols[-1:] + cols[:-1]]
    year_df["color"] = pd.Series(["#ac5430",
                                  "#ebb783",
                                  "#ebb783",
                                  "#499092",
                                  "#499092",
                                  "#ac5430",
                                  "#ac5430",
                                  "#499092",
                                  "#b77745",
                                  "#b77745"])
    year_df.to_csv("{}.csv".format(year), index=False)
```
